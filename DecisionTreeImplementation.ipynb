```python
#Seperate data as features and label
features = data.iloc[:,0:2].values
label = data.iloc[:,2].values
```

```python
SL = 0.05
CL = 1-SL

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

for rs in range(1,300):
  X_train,X_test,y_train,y_test = train_test_split(features,
                                                   label,
                                                   test_size=0.2,
                                                   random_state=rs)

  model = DecisionTreeClassifier()
  model.fit(X_train,y_train)
  trainScore = model.score(X_train,y_train)
  testScore = model.score(X_test,y_test)
  bestTestScore = 0

  #if testScore > trainScore:
  print(f"Test Score : {testScore} TrainScore: {trainScore} RS: {rs}")
```

```python
#As above result, it is EVIDENT that Decision Tree tend to OVERFIT in default mode.
#Overfitting means
#  1. Model performs BEST with known data, but performs decent to pathetic with UNKNOWN DATA.
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#Note: If the label column has missing value, simply delete that record
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.describe()
```

```python
#Seperate data as featues and label
features = data.iloc[:,[0]].values
label = data.iloc[:,[1]].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor


CL = 0.7
for rs in range(1,300):
  X_train,X_test,y_train,y_test = train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=rs)

  for depth in range(1,10):
    model = DecisionTreeRegressor(max_depth=depth)
    model.fit(X_train,y_train)

    testScore = model.score(X_test,y_test)
    trainScore = model.score(X_train,y_train)

    if testScore > trainScore and testScore >= CL:
      print(f"Test Score {testScore} and Train score {trainScore} and RandomState {rs} and Depth {depth}")
```
